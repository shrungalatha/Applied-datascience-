```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
def bar_graph1():
    """Bar graph showing number of restaurants in India across diffirent cities. 
    
    """
    sns.set_style("whitegrid")
    plt.figure(figsize=(10, 6)) 

    city_counts = data_res['City'].value_counts() 
    sns.barplot(x=city_counts.index, y=city_counts.values, palette="viridis") 
    plt.xticks(rotation=45, ha="right") 
    plt.xlabel('City') 
    plt.ylabel('Number of Restaurants') 
    plt.title('Number of Restaurants in Each City') 
    plt.tight_layout() 
    plt.show()
    return()
```

```python
def bar_graph2():
    # Calculating the average cost per city assigned to average_cost_per_city. 
    average_cost_per_city = data_res.groupby('City')['Cost'].mean().sort_values(ascending=False)

    # Ploting the average cost per city

    plt.figure(figsize=(12, 8))
    average_cost_per_city.plot(kind='bar', color='Red')
    plt.title('Average Restaurant Cost per City' , fontsize=18)
    plt.xlabel('City' , fontsize=18)
    plt.ylabel('Average Cost(INR)' , fontsize=18)
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout() 
    plt.gca().invert_xaxis()
    plt.show()
    return()
```

```python
def bar_graph3():
    # Count the number of outlets for each restaurant
    outlet_counts = data_res['Name'].value_counts()

    # Get the top 10 restaurants with the most outlets
    top_10_restaurants = outlet_counts.head(10)

    # Plot the horizontal bar chart
    plt.figure(figsize=(14, 7))
    top_10_restaurants.plot(kind='barh', color=plt.cm.Purples_r(range(len(top_10_restaurants))))
    plt.title('Top 10 Restaurants With Maximum Outlets' , fontsize=18)
    plt.xlabel('Outlets' , fontsize=18)
    plt.ylabel('Restaurant Name' , fontsize=18)
    plt.gca().invert_yaxis() 
    plt.tight_layout()
    plt.show()
```

```python
def scatter_plot():
    "scatter plot showing the relation between ratings and votes"
    plt.figure(figsize=(10, 6)) 
    sns.scatterplot(x='Rating', y='Votes', data=data_res,hue='Rating', alpha=0.6, edgecolor=None, palette="viridis") 
    plt.xlabel('Rating') 
    plt.ylabel('Votes') 
    plt.title('Relationship Between Rating and Votes') 
    plt.legend(loc='upper left',bbox_to_anchor=(1,1))
    plt.show()
```

```python
def heatmap():
    "the heatmap explaining corelation between rating, cost and votes"
    plt.figure(figsize=(8, 6)) 

    correlation_matrix = data_res[['Rating','Cost','Votes']].corr()
    sns.heatmap(correlation_matrix, annot=True, cmap="viridis", cbar_kws={'label': 'Correlation Coefficient'}) 
    plt.title('Correlation Matrix Heatmap') 
    plt.show()
```
